```python
from PyPDF2 import PdfReader
from langchain.chat_models import ChatOpenAI
from langchain.chains import load_summarize_chain
import tiktoken

# Step 1: Extract text from the PDF
file_path = '/content/Raw_Data (1).pdf'
pdfreader = PdfReader(file_path)

text = ''
for page in pdfreader.pages:
    content = page.extract_text()
    if content:
        text += content

# Step 2: Tokenize the text
encoding = tiktoken.encoding_for_model("gpt-3.5-turbo")
tokens = encoding.encode(text)

print(f"Original token count: {len(tokens)}")

# Step 3: Check if the token count exceeds the limit
if len(tokens) > 4096:
    # Initialize the ChatOpenAI model
    llm = ChatOpenAI(temperature=0, model_name='gpt-3.5-turbo')

    # Create the summarize chain
    chain = load_summarize_chain(llm, chain_type='map_reduce', verbose=False)

    # Split the text into manageable chunks
    chunk_size = 2000  # Adjust as necessary
    chunks = [text[i:i + chunk_size] for i in range(0, len(text), chunk_size)]

    # Step 4: Summarize using the chain
    summary = chain.run(chunks)
    print(f"Summarized token count: {len(encoding.encode(summary))}")

else:
    summary = text  # No need to summarize if within limits

# Print the summarized text
print("Summary:")
print(summary)
```
